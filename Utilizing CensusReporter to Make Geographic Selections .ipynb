```python
# -*- coding: utf-8 -*-
"""
Created on Wed Oct 10 18:56:37 2018

@author: dcbs1
"""

import pandas as pd
import numpy as np
import requests
import time

data = pd.read_csv("C:\\Users\\dcbs1\\Desktop\\Sample\\msas.csv")
#print(data)
'''
tableids = ['B01001', 'B03002', 'B19001']
geo_id = '31000US'+'15380'
tableidstring = ','.join(tableids)
requesturl = 'http://api.censusreporter.org/1.0/data/show/latest?table_ids=%s&geo_ids=%s' % (tableidstring, geo_id)
print(requesturl)
loadedjson = requests.get(requesturl)
print(loadedjson.json())

'''
age_groups = ['007','008','009','010','011','012','031','032','033','034','035','036']
hispanics =['012']
income_group = ['011', '012']

max_id_age = max_age_sum = 0
max_id_hispanic = max_hispanic_sum = 0
max_id_incomegroup = max_incomegroup_sum =  0
max_mean_percent = max_mean_percent_id = 0

tableids = ['B01001', 'B03002', 'B19001']
geo_id = ['31000US'+str(city_id) for city_id in data['CBSA'].values]
tableidstring = ','.join(tableids)

for city_id in geo_id:
    print(city_id)
    requesturl = 'http://api.censusreporter.org/1.0/data/show/latest?table_ids=%s&geo_ids=%s' % (tableidstring, city_id)
    loadedjson = requests.get(requesturl)
    if loadedjson.status_code!= 200:
        time.sleep(5)
        loadedjson = requests.get(requesturl)
        
    loadedjson_dict = loadedjson.json()
    sum_of_age_group = 0
    for age_id in age_groups:
        sum_of_age_group = sum_of_age_group + loadedjson_dict['data'][city_id]['B01001']['estimate']['B01001'+age_id]
    
    sum_of_hispanic = loadedjson_dict['data'][city_id]['B03002']['estimate']['B03002012']
    if sum_of_hispanic is None:
        sum_of_hispanic = 0
    
    sum_of_incomegroup = 0
    for in_id in income_group:
        sum_of_incomegroup = sum_of_incomegroup + loadedjson_dict['data'][city_id]['B19001']['estimate']['B19001'+in_id]
    
    if sum_of_age_group > max_age_sum:
        max_age_sum = sum_of_age_group
        max_id_age = city_id
        
    if sum_of_hispanic > max_hispanic_sum:
        max_hispanic_sum = sum_of_hispanic
        max_id_hispanic = city_id
        
    if sum_of_incomegroup > max_incomegroup_sum:
        max_incomegroup_sum = sum_of_incomegroup
        max_id_incomegroup = city_id 
        
    if loadedjson_dict['data'][city_id]['B03002']['estimate']['B03002001'] is None:
        mean_percent = (sum_of_age_group/loadedjson_dict['data'][city_id]['B01001']['estimate']['B01001001']+
                    sum_of_incomegroup/loadedjson_dict['data'][city_id]['B19001']['estimate']['B19001001'])*100/3
    else:  
        mean_percent = (sum_of_age_group/loadedjson_dict['data'][city_id]['B01001']['estimate']['B01001001']+
                    sum_of_hispanic/loadedjson_dict['data'][city_id]['B03002']['estimate']['B03002001']+
                    sum_of_incomegroup/loadedjson_dict['data'][city_id]['B19001']['estimate']['B19001001'])*100/3

    if mean_percent > max_mean_percent:
        max_mean_percent = mean_percent
        max_mean_percent_id = city_id
    
    
dict_id_city = dict(zip(data['CBSA'].values, data['NAME'].values))
print("City with most number of people within age-group of 18-34: ", dict_id_city[int(max_id_age[7:])], '. Count : ', max_age_sum)
print("City with most Number of Hispanics : ", dict_id_city[int(max_id_hispanic[7:])], '. Count : ', max_hispanic_sum)
print("City with most Number of Income-group : ", dict_id_city[int(max_id_incomegroup[7:])], '. Count : ', max_incomegroup_sum)
print("City with most Number of Mean-percent : ", dict_id_city[int(max_mean_percent_id[7:])], '. Count : ', max_mean_percent)
```
