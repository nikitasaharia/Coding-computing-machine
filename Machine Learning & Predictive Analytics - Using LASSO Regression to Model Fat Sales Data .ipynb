```python
# -*- coding: utf-8 -*-
"""
Created on Tue Oct  2 13:51:05 2018

@author: dcbs1
"""

import pandas as pd
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LassoLarsCV
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore', 'Solver terminated early.*')

alldata = pd.read_csv("C:\\Users\\dcbs1\\Desktop\\Sample\\ratios.csv")
#DEleting predictor variables
allvariablenames = list(alldata.columns.values)
alldata_np = alldata.values
data_frame_np = np.delete(alldata_np, [1,2,3,4,5,6,7], 1)
allvariablenames = np.delete(allvariablenames, [1,2,3,4,5,6,7])
#print(data_frame_np)


#load predictors into dataframe
predictors = alldata[allvariablenames[1:]]  

#load target into dataframe
target = alldata['# Purchases']   

#print(len(allvariablenames)) #183 variables inclusind the y variable



# split data into train and test sets, with 30% retained for test

pred_train, pred_test, tar_train, tar_test = train_test_split(predictors, target, test_size=.3, random_state=123)    

#we now have four new pandas data frames. (1) pred_train, the predictors training set (2) pred_test, the predictors test test (3) target_train, the target training set and (4) tar_test, the target test set. We'll feed these datasets to our model in the next step.
#Now the hard part, where I'm looking at you to correctly use python documentation. Use LassoLarsCV (Links to an external site.)Links to an external site. to build a LASSO model. Here are the parameters you'll need to include in your model. I'm looking for you to implement it correctly. 
#Use LassoLarsCV and feed the following parameters in:
#cv = 10 (this performs a 10-fold cross validation, to make sure our results aren't do to random ordering of the data)
#precompute=False (not necessary)
#name this model 'model'.

#Next, we need to fit our newly created model with your training data. To do this, use .fit() on your newly created model object (Links to an external site.)Links to an external site.. Remember what X and y mean from doing regression?
#Congrats. You've just built your first LASSO model. Next, let's see what variables were significant. Next, let's extract the coefficients from the model: 

# Lassolar 
 
model = LassoLarsCV(cv=10, max_iter=100000).fit(pred_train, tar_train)
model.score(pred_train, tar_train) 


#build coefficent chart

predictors_model=pd.DataFrame(allvariablenames[1:])
predictors_model.columns = ['label']
predictors_model['coeff'] = model.coef_



#Finding the mean standard error
for index, row in predictors_model.iterrows():
    if row['coeff'] > 0:
        print(row.values)
        
        
#calculating mean squared difference for question pred_train and tar_train
model_1 = model.predict(pred_train)
mse = np.mean((model_1 - tar_train)**2)
print(mse)


#calculating mean squared for pred_test and tar_test
model_2 = model.predict(pred_test)
mse1 = np.mean((model_2 - tar_test)**2)
print(mse1)
        

#Calculating r squared for training sets
#r squared
rsquared_train=model.score(pred_train,tar_train)
print ('training data R-square')
print(rsquared_train)

#Calculating the r squared for test sets
rsquared_train1=model.score(pred_test,tar_test)
print ('test data R-square')
print(rsquared_train1)


#Print the y intercept
print("y interecept:")
print(model.intercept_)
```
